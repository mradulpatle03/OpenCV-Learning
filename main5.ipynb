```python
import cv2

cap = cv2.VideoCapture(0)  # 0 = default camera

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    cv2.imshow("Video", frame)
    
    if cv2.waitKey(1) & 0xFF == 27:  # ESC to exit
        break
fps = cap.get(cv2.CAP_PROP_FPS)
print("FPS:", fps)
cap.release()
cv2.destroyAllWindows()
```

```python
# optimization : skip some frames
frame_count = 0
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    
    if frame_count % 5 != 0:
        continue   # skip this frame

    cv2.imshow("Processed Frame", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# saving the video 

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    out.write(frame)  # save frame

    cv2.imshow("Video", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
# multi camera input 

cap1 = cv2.VideoCapture(0)
cap2 = cv2.VideoCapture(1)

while True:
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()

    if ret1:
        cv2.imshow("Cam 1", frame1)
    if ret2:
        cv2.imshow("Cam 2", frame2)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap1.release()
cap2.release()
cv2.destroyAllWindows()
```

```python
# Video buffering = storing frames temporarily before processing

# Why needed?
# Handle slow processing
# Avoid frame drops
# Used in streaming apps
```

```python
from collections import deque
import cv2
buffer = deque(maxlen=10)
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    buffer.append(frame)

    # process latest frame
    latest_frame = buffer[-1]

    cv2.imshow("Buffered Frame", latest_frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break
```

```python
# geometric vision: Transforming images (changing perspective, alignment, shape)

# What is Homography?
# A transformation that maps:
# one plane → another plane
```

```python
import numpy as np
import cv2
img = cv2.imread("test.jpg")
pts_src = np.array([[100,100], [400,100], [400,400], [100,400]])
pts_dst = np.array([[0,0], [300,0], [300,300], [0,300]])

matrix, _ = cv2.findHomography(pts_src, pts_dst)

warped = cv2.warpPerspective(img, matrix, (300,300))
cv2.imshow("Warped", warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Perspective Correction : 
# Camera captures images with distortion (tilted view)

# Goal
# Make it look like:
# Shot from directly above
```

```python
import cv2
import numpy as np

# Step 1: Load image
image = cv2.imread("doc.jpg")

# Step 2: Define 4 corner points (clockwise order)
# (Top-left, Top-right, Bottom-right, Bottom-left)
pts_src = np.array([
    [100, 150],   # Top-left
    [400, 130],   # Top-right
    [420, 300],   # Bottom-right
    [80, 320]     # Bottom-left
], dtype="float32")

# Step 3: Define destination points (rectangle)
width = 300
height = 200

pts_dst = np.array([
    [0, 0],
    [width - 1, 0],
    [width - 1, height - 1],
    [0, height - 1]
], dtype="float32")

# Step 4: Compute transformation matrix
matrix = cv2.getPerspectiveTransform(pts_src, pts_dst)

# Step 5: Apply warp
warped = cv2.warpPerspective(image, matrix, (width, height))

# Step 6: Show results
cv2.imshow("Original", image)
cv2.imshow("Warped", warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image alignment:Align two images so they match perfectly

img1 = cv2.imread("test.jpg")
img2 = cv2.imread("test.jpg")

orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)
matches = sorted(matches, key=lambda x: x.distance)
img_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], None, flags=2)
cv2.imshow("Matches", img_matches)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
# panorama stitching: Combine multiple images into a single panorama

img1 = cv2.imread("test.jpg") 
img2 = cv2.imread("test.jpg") 
stitcher = cv2.Stitcher_create() 
status, pano = stitcher.stitch([img1, img2]) 
if status == cv2.Stitcher_OK: 
    cv2.imshow("Panorama", pano) 
else: print("Error during stitching:", status) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
# Watershed algorithm : 
import numpy as np
img = cv2.imread("shapes.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

sure_bg = cv2.dilate(opening, kernel, iterations=3)

dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
_, sure_fg = cv2.threshold(dist_transform, 0.7*dist_transform.max(), 255, 0)

sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

markers = cv2.connectedComponents(sure_fg)[1]
markers = markers + 1
markers[unknown == 255] = 0

markers = cv2.watershed(img, markers)

img[markers == -1] = [255,0,0]  # boundaries
cv2.imshow("Watershed", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# using grabcut algorithm for background removal
import cv2
import numpy as np

# Step 1: Load Image
img = cv2.imread("virat.jpg")

if img is None:
    print("Error: Image not found")
    exit()

img = cv2.resize(img, (500, 600))

# Step 2: Create initial mask
mask = np.zeros(img.shape[:2], np.uint8)

# Step 3: Create background & foreground models
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

# Step 4: Define rectangle
rect = (50, 50, 400, 500)

# Step 5: Apply GrabCut
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

# Step 6: Refine mask
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype("uint8")

# Step 7: Extract foreground
result = img * mask2[:, :, np.newaxis]

# Step 8: Create white background output
white_bg = np.ones_like(img, dtype=np.uint8) * 255
final = np.where(mask2[:, :, np.newaxis] == 1, img, white_bg)

# Step 9: Show output
cv2.imshow("Original", img)
cv2.imshow("Foreground Only", result)
cv2.imshow("White Background", final)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

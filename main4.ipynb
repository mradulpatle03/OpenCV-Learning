```python
# Moving object detection using background subtraction
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

fgbg = cv2.createBackgroundSubtractorMOG2()

kernel = np.ones((3,3), np.uint8)

while True:
    ret, frame = cap.read()
    
    mask = fgbg.apply(frame)
    
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    contours, _ = cv2.findContours(
        mask,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )
    
    for cnt in contours:
        if cv2.contourArea(cnt) > 500:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
    
    cv2.imshow("Frame", frame)
    cv2.imshow("Mask", mask)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Text detection and recognition using OpenCV and Tesseract OCR(Optical Character Recognition)
# scanning documents, reading number plates, or extracting text from invoices.



import cv2

img = cv2.imread("test.jpg")
# step 1: grayscale conversion
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow("Gray", gray)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# step2 : binarization
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
# step3 : noise removal
import numpy as np

kernel = np.ones((3,3), np.uint8)
clean = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
```

```python
# Optional: Dilation / Erosion
# Dilation thickens text, erosion thins it. Useful for broken letters.

# Text Detection: Find where the text is (bounding boxes).
# Text Recognition: Read what the text actually says.
```

```python
import pytesseract

text = pytesseract.image_to_string(gray)
print("Detected Text:")
print(text)

h, w, _ = img.shape
boxes = pytesseract.image_to_boxes(gray)
for b in boxes.splitlines():
    b = b.split()
    x, y, w_box, h_box = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    cv2.rectangle(img, (x, h - y), (w_box, h - y + h_box), (0,255,0), 2)

cv2.imshow("Text Boxes", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

text = pytesseract.image_to_string(gray, lang='eng')
```

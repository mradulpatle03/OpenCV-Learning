```python
import cv2
import numpy as np

img = cv2.imread("test.png")
template = cv2.imread("test1.png")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

# Edge-based matching (better)
img_edge = cv2.Canny(img_gray, 50, 150)
template_edge = cv2.Canny(template_gray, 50, 150)

result = cv2.matchTemplate(img_edge, template_edge, cv2.TM_CCOEFF_NORMED)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

print("Max similarity:", max_val)

h, w = template_gray.shape[:2]
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(img, top_left, bottom_right, (0,255,0), 2)

cv2.imshow("Detected", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Pose estimation = Detecting human body joints (keypoints) from an image/video
import cv2
import mediapipe as mp

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

mp_draw = mp.solutions.drawing_utils

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert BGR → RGB
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process frame
    result = pose.process(rgb)

    # Draw landmarks
    if result.pose_landmarks:
        mp_draw.draw_landmarks(
            frame,
            result.pose_landmarks,
            mp_pose.POSE_CONNECTIONS
        )

    cv2.imshow("Pose Estimation", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```

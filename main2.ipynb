<!-- Edge Detection -->

```python
import cv2

img = cv2.imread('test.jpg')
```

```python
# sobel operator
# Detects edges using gradients
# (1,0) → detect horizontal change
# (0,1) → detect vertical change

sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1)

sobel = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

# cv2.imshow('Sobel', sobel)
# cv2.waitKey(0)  
# cv2.destroyAllWindows()
```

```python
# laplacian operator
# detects edges in all directions

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# cv2.imshow('Laplacian', laplacian)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

<!-- Thresholding -->

```python
# Only 2 values → 0 or 255
# Binary image

ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# cv2.imshow('Threshold', thresh)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

<!-- Types:
1) cv2.THRESH_BINARY : simple
2) cv2.THRESH_BINARY_INV : Black becomes white, white becomes black
3) cv2.THRESH_TRUNC : Values above threshold become threshold
4) cv2.THRESH_TOZERO : Below threshold → 0 -->

```python
# adaptive thresholding
# Threshold value is calculated for smaller regions
# More accurate for images with varying illumination

thresh = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)
# cv2.imshow('Adaptive Threshold', thresh)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# types:
# Mean : ADAPTIVE_THRESH_MEAN_C
# Gaussian : ADAPTIVE_THRESH_GAUSSIAN_C
```

<!-- Morphological operations -->

```python
# Goal
# Remove noise
# Fix broken shapes
# Improve object detection
import numpy as np
kernel = np.ones((3,3), np.uint8)


erosion = cv2.erode(thresh, kernel, iterations=1)
# Removes white pixels at boundaries
# Eliminates small noise

dilation = cv2.dilate(thresh, kernel, iterations=1)
# Adds pixels to boundaries
# Fills small gaps
```

```python
# Opening (Erosion → Dilation)
# Removes noise but keeps object intact
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Closing (Dilation → Erosion)
# Fill gaps in object
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
# Morphological Gradient
# Extract edges using morphology

gradient = cv2.morphologyEx(thresh, cv2.MORPH_GRADIENT, kernel)
# Difference between dilation and erosion
```

```python
# Erosion	    Remove noise
# Dilation	    Fill gaps
# Opening	    Remove small objects
# Closing	    Fill holes
# Gradient	    Get edges
```
